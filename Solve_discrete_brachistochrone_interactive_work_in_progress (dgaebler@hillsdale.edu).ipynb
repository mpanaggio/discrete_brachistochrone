```python
a=10 # x-coordinate for endpoint
b=-1  # y-coordinate for endpoint
N=4 # number of segments
```

```python
def angles_with_previous(x,y):
    angles=[]
    for k in range(1,len(x)-1):
        vec_prev=np.array([x[k]-x[k-1],y[k]-y[k-1]])
        vec_next=np.array([x[k+1]-x[k],y[k+1]-y[k]])
        #vec_prev=vec_prev/np.linalg.norm(vec_prev)
        #vec_next=vec_next/np.linalg.norm(vec_next)
        angles.append(np.arccos(np.dot(vec_prev,vec_next)/np.sqrt(np.dot(vec_prev,vec_prev)*np.dot(vec_next,vec_next))))
    return angles
```
